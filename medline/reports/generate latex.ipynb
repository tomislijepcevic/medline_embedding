```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table('results/mean.tsv', names=['model', 'povprečna točnost'])
```

```python
df = pd.read_table('results/categories.tsv', index_col=0)
df.index.name = 'skupina izrazov'
```

```python
for row in df.iterrows():
    k = row[0]
    v = list(*row[1:])
    i = np.argmax(v)
    v = ['%.3f' % f for f in v]
    v[i] = r'\textbf{%s}' % v[i]
    l = [k] + v

    print(' & '.join(l) + r'\\ \n')
```

```python
categories = [l.rstrip('\n').split('\t')[1]
              for l in open('categories.tsv')]
```

```python
for category in categories:
    try:
        df = pd.read_table('results/categories/%s.tsv' % category, index_col=0)
    except:
        continue
        
    df = df.rename(columns={'support': 'primerov'})
    df.index.name = 'izraz'

    for row in df.iterrows():
        l = [row[0], str(int(row[1][0]))]
        r = list(row[1][1:])
        i = np.argmax(r)
        r = ['%.3f' % f for f in r]
        r[i] = r'\textbf{%s}' % r[i]
        
        print(' & '.join(l + r) + r'\\ \n')

    print()
```
